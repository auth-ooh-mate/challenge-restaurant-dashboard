```python
import os

import numpy as np
import pandas as pd
import plotly.express as plt

DATA_FOLDER = os.path.abspath('data')
DATASET_FILE = os.path.join(DATA_FOLDER, 'TA_restaurants_curated.csv')

PROCESSED_DATA_FOLDER = os.path.join(DATA_FOLDER, 'processed')

CITIES_FILE = os.path.join(PROCESSED_DATA_FOLDER, 'cities.csv')
STYLES_FILE = os.path.join(PROCESSED_DATA_FOLDER, 'cuisine_styles.csv')
CITY_STYLES_FILE = os.path.join(PROCESSED_DATA_FOLDER, 'city_cuisine_styles.csv')
```

```python
# load dataframe from csv file
def load_df(file):

    if not os.path.exists(file): 
        print('File doesn\'t exist:', file)
        return

    df = pd.read_csv(file)
    return df
```

```python
cities_df = load_df(CITIES_FILE)
cities_all = cities_df.loc[cities_df['city_name']=='ALL']
cities_df.drop(cities_df.loc[cities_df['city_name']=='ALL'].index, inplace=True)

cities_df.sort_values(by=['total_cuisine_styles'], ascending=False, inplace=True)

cities_df.head()
```

```python
cities_styles_df = load_df(CITY_STYLES_FILE)
cities_styles_all = cities_styles_df.loc[cities_styles_df['city_name']=='ALL']
cities_styles_df.drop(cities_styles_df.loc[cities_styles_df['city_name']=='ALL'].index, inplace=True)

cities_styles_all = cities_styles_all.sort_values(by=['total_restaurants'], ascending=False)

cities_styles_all
```
